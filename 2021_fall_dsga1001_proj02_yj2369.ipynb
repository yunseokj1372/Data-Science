### Dataset description

This dataset features ratings data of 400 movies from 1097 research participants. 

* 1st row: Headers (Movie titles/questions) â€“ note that the indexing in this list is from 1
* Row 2-1098: Responses from individual participants
* Columns 1-400: These columns contain the ratings for the 400 movies (0 to 4, and missing)
* Columns 401-421: These columns contain self-assessments on sensation seeking behaviors (1-5)
* Columns 422-464: These columns contain responses to personality questions (1-5)
* Columns 465-474: These columns contain self-reported movie experience ratings (1-5)
* Column 475: Gender identity (1 = female, 2 = male, 3 = self-described)
* Column 476: Only child (1 = yes, 0 = no, -1 = no response)
* Column 477: Movies are best enjoyed alone (1 = yes, 0 = no, -1 = no response)

Note that we did most of the data munging for you already (e.g. Python interprets commas in a csv file as separators, so we removed all commas from movie titles), but you still need to handle missing data.

In this problem, under **the most correlated**, we consider the largest correlation in the absolute value.


1.1. For every user in the given data, find its most correlated user. 

1.2. What is the pair of the most correlated users in the data? 

1.3. What is the value of this highest correlation?

1.4. For users 0, 1, 2, \dots, 9, print their most correlated users. 



```python
# import csv
df = np.genfromtxt('.\movieReplicationSet.csv', delimiter = ',', skip_header = 1)
dfp = pd.read_csv('.\movieReplicationSet.csv')
dfp.columns
```

```python
mov = df[:,:400]
# assigning null values average
for i in range(mov.shape[1]):
    temp = np.isnan(mov[:,i])
    cmean = np.nanmean(mov[:,i])
    mov[temp,i] = cmean
```

```python
np.max(mov)
```

```python
# 1.1
CM= np.corrcoef(mov)
mCM = abs(CM) # absolute value
# fill diagonal with 0 so as to get the correlation between user
np.fill_diagonal(mCM,0)
for j in range(len(mCM)):
    print('User', j, ' is correlated most with user', np.argmax(mCM[:,j]), np.max(mCM[:,j]))
```

```python
# 1.2
maxcorruser = np.unravel_index(np.argmax(mCM, axis=None), mCM.shape)
print('The pair of most correlated users are',maxcorruser)
```

```python
# 1.3
print('The maximum correlation is', mCM[maxcorruser])
```

```python
# 1.4 
for j in range(10):
    print('User', j, ' is correlated most with user', np.argmax(mCM[:10,j]))
```

```python
# Just the 10 users most correlated
for j in range(10):
    print('User', j, ' is correlated most with user', np.argmax(mCM[:,j]), np.max(mCM[:,j]))
```

```python
# Part 1
# Setting initial Dataset

movie_set = df
# assigning null values average
for i in range(movie_set.shape[1]):
    temp = np.isnan(movie_set[:,i])
    cmean = np.nanmean(movie_set[:,i])
    movie_set[temp,i] = cmean
    
    
df_rate = movie_set[:,:400]
df_pers = movie_set[:,400:474]
```

```python
# Part 2
# splitting according to 0.8:0.2 ratio of train and test
X_train, X_test, y_train, y_test = train_test_split(df_rate, df_pers, test_size=0.2)

```

```python
# 2.1
# linear regression
regr  = LinearRegression()
regr.fit(X_train,y_train)
betas = regr.coef_ # m
yInt = regr.intercept_  # b

# prediction
trainp = regr.predict(X_train)
testp = regr.predict(X_test)

# mean_squared_error
# train_error = np.empty(np.shape(trainp)[1])
# test_error = np.empty(np.shape(trainp)[1])


# for i in range(np.shape(trainp)[1]):
#     train_error[i] = mean_squared_error(y_train[:,i], trainp[:,i] )
#     test_error[i] = mean_squared_error(y_test[:,i], testp[:,i])

train_error = mean_squared_error(y_train, trainp )
test_error = mean_squared_error(y_test, testp)


print(train_error)
print(test_error)
```

```python
# Alternative Way for each column
train_mse_array = np.zeros(74)
test_mse_array = np.zeros(74)

for i in range(74):
    xx_train, xx_test, yy_train, yy_test = train_test_split(df_rate, df_pers[:,i], test_size=0.2)
    yy_train = np.reshape(yy_train,[np.shape(yy_train)[0],1])
    regr  = LinearRegression()
    regr.fit(xx_train,yy_train)
    train_mse_array[i] = mean_squared_error(yy_train, regr.predict(xx_train) )
    test_mse_array[i]= mean_squared_error(yy_test, regr.predict(xx_test))
    
print(train_mse_array)
print(test_mse_array)

print(np.sum(train_mse_array))
print(np.sum(test_mse_array))
```

```python
# 2.2

alpha = [0.0, 1e-8, 1e-5, 0.1, 1, 10]

trainr_error = np.empty(len(alpha))
testr_error = np.empty(len(alpha))

for lamb in range(len(alpha)):
    ridg = Ridge(alpha = alpha[lamb])
    ridg.fit(X_train, y_train)
    
    trainr = ridg.predict(X_train)
    testr = ridg.predict(X_test)
    
#     for i in range(np.shape(trainp)[1]):
#         trainr_error[lamb,i] = mean_squared_error(y_train[:,i], trainr[:,i] )
#         testr_error[lamb,i] = mean_squared_error(y_test[:,i], testr[:,i])
    
    trainr_error[lamb] = mean_squared_error(y_train, trainr )
    testr_error[lamb] = mean_squared_error(y_test, testr)

    
print("Train Error: ", trainr_error)
print("Test Error: ", testr_error)
# minimum
minerr_ridge = np.min(testr_error)

opt_alpha_train = alpha[np.argmin(trainr_error)]
opt_alpha_test = alpha[np.argmin(testr_error)]

print("Optimal train alpha: ", opt_alpha_train)
print("Optimal test alpha: ", opt_alpha_test)
```
