## Data analysis project 1:
## Hypothesis testing of movie ratings data

## Dataset description: 
This dataset features ratings data of 400 movies from 1097 research participants.

1st row: Headers (Movie titles/questions) â note that the indexing in this list is from 1 Row 2-1098: Responses from individual participants

Columns 1-400: These columns contain the ratings for the 400 movies (0 to 4, and missing)
    
Columns 401-421: These columns contain self-assessments on sensation seeking behaviors (1-5)
    
Columns 422-464: These columns contain responses to personality questions (1-5)
    
Columns 465-474: These columns contain self-reported movie experience ratings (1-5)
    
Column 475: Gender identity (1 = female, 2 = male, 3 = self-described)
    
Column 476: Only child (1 = yes, 0 = no, -1 = no response)
    
Column 477: Movies are best enjoyed alone (1 = yes, 0 = no, -1 = no response)
    

```python
import numpy as np
import pandas as pd
from scipy import stats

df1 = np.genfromtxt('movieReplicationSet.csv', delimiter= ',', skip_header = 1)
df2 = pd.read_csv('movieReplicationSet.csv')
df3 = np.genfromtxt('movieReplicationSet.csv', delimiter= ',', names = True)
df3
```

1) Are movies that are more popular (operationalized as having more ratings) rated higher than movies that are less popular?

```python
q1 = df1[:,:400]
q1
```

```python
temp = np.isnan(q1)
temp2 = temp*1 
temp3 = np.sum(temp2,axis = 0)

temp
```

```python
M = np.median(temp3)
l1 = []
l2 = []
l11 = []
l22 = []

t1 = np.where(temp3<M) # more popular movie since have more nan
t2 = np.where(temp3>=M)
for i,j in zip(t1[0],t2[0]):
    temp1 = q1[:,i][np.isfinite(q1[:,i])]
    temp2 = q1[:,j][np.isfinite(q1[:,j])]
    l1.append(np.median(temp1))
    l2.append(np.median(temp2))
    l11.append(np.average(temp1))
    l22.append(np.average(temp2))
M
```

```python
u1,p1 = stats.mannwhitneyu(l1,l2, alternative = 'greater')
u2,p2 = stats.mannwhitneyu(l11,l22, alternative = 'greater')

print(u1,p1,'\n',u2,p2)
```

```python
t1 = np.where(temp3<M)
t2 = np.where(temp3>=M)
ch = []
for i in t1[0]:
    temp1 = q1[:,i][np.isfinite(q1[:,i])]
    for j in t2[0]:
        temp2 = q1[:,j][np.isfinite(q1[:,j])]
        t, p = stats.mannwhitneyu(temp1, temp2, alternative= 'greater')
        ch.append(p)
ch
```

```python
check = np.array([k < 0.005 for k in ch]) *1
tot = len(check)
popular = sum(check)
unpopular = tot-popular
print(popular, unpopular)
```

```python
newtemp = [0.005]*len(check)
newcheck = np.array(ch)
stats.ttest_ind(newcheck, newtemp, alternative = 'less')
```
